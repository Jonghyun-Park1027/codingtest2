```python
def factors(x):
    data = []
    for i in range(1, x):
        if x % i == 0:
            data.append(i)
    return data


while True:
    n = int(input())
    if n == -1:
        break
    data = factors(n)
    # print(data)
    if sum(data) == n :
        data = " + ".join([str(i) for i in data])
        print(f"{n} = {data}")
    else:
        print(f"{n} is NOT perfect.")
```

```python
n, b = input().split()

# 진수의 자리이동에 따른 제곱수를 구해야한다
count = 0
# answer를 10진수로 더해서 담을 값이 있어야 한다
answer = 0
# str을 for문으로 돌린다, 근데 거꾸로해야함
for i in n[::-1]:
    # ord('A') 는 65 ord(n)-55를 하면 된다 카운트는 지수의 제곱수가 된다
    # 각 숫자에 제곱수를 해주고 answer에 누적합시킨다.
    # 근데 숫자가 아닐경우만 통과시켜야함
    if ord(i) - 55 >= 10:
        answer += ((ord(i) - 55) * (int(b) ** count))
    # 아닐경우 그냥 숫자를 넣어야함
    else:
        answer += int(i) * (int(b) ** count)
    # 1회 돌때마다 count를 1씩 증가시켜준다
    count += 1
    # print(i)
# answer를 출력한다
print(answer)
```

```python
#왕개미
graph = []
for _ in range(10):
    graph.append(list(map(int, input().split())))

start_x = 1
start_y = 1

graph[start_x][start_y] = 9
# 미로 상자의 구조가 0(갈 수 있는 곳), 1(벽 또는 장애물)로 주어지고,
# 먹이가 2로 주어질 때, 성실한 개미의 이동 경로를 예상해보자.
while True:

    if start_x < 8:
        if graph[start_x][start_y + 1] == 0:

            start_y += 1
            graph[start_x][start_y] = 9
        elif graph[start_x][start_y + 1] == 2:

            start_y += 1
            graph[start_x][start_y] = 9
            break
        elif graph[start_x][start_y + 1] == 1:
            pass
    elif start_y < 8:
        if graph[start_x + 1][start_y] == 0:

            start_x += 1
            graph[start_x][start_y] = 9
        elif graph[start_x + 1][start_y] == 2:

            start_x += 1
            graph[start_x][start_y] = 9
            break
        elif graph[start_x + 1][start_y] == 1:
            pass
    # 만약에 가장 아래이고 오른쪽이면 못가 그럼 멈춰춰
    elif start_x == 8:
        if graph[start_x][start_y + 1] == 0:

            start_y += 1
            graph[start_x][start_y] = 9
        elif graph[start_x][start_y + 1] == 2:

            start_y += 1
            graph[start_x][start_y] = 9
            break
        if start_x == 8 and start_y == 8:
            break
    elif start_y == 8:
        if graph[start_x + 1][start_y] == 0:

            start_x += 1
            graph[start_x][start_y] = 9
        elif graph[start_x + 1][start_y] == 2:

            start_x += 1
            graph[start_x][start_y] = 9
            break
        if start_x == 8 and start_y == 8:
            break
    # 오른쪽이랑 아래가 둘다 막혀있으면 못가가
    elif graph[start_x + 1][start_y] == 1 and graph[start_x][start_y + 1] == 1:
        # else:
        break

    for i in graph:
        print(*i)
```
